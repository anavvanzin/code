# Install the ICONCLASS Python package

Initialize the library

```python
from iconclass import init

ic = init()
```

Lookup a notation

```python
a = ic["53A2"]
```

Fetch the texts and representations

```python
print(a)
print(repr(a))
print(a())
print(a('de'))

```

Show the children of a notation

```python
for child in a:
    print(child('it'))
```

Show the hierarchy

```python
for p in a.path():
    print(repr(p))
```
